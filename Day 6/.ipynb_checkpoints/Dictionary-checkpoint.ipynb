## Dictionary 
- key and value format


```python
person={
    "name":"Ram",
    "address":"KTM",
    "age":34,
    "salary":56000.4,
    "is_alive":True
}
print(person)
```

```python
print(type(person))
```

```python
print(person.keys())
```

```python
print(person.values())
```

```python
print(person.items())
```

```python
print(person.get('name'))
print(person.get('age'))
print(person.get('phone'))# none because phone key is not in dect
print(person.get('is_alive'))
print(person.get('salary'))
print(person.get('address'))
```

```python
person.pop('age')
```

```python
print(person)
```

```python
person["is_alive"]=False
print(person)
```

```python
x=person.copy()
print(x)
```

```python
x.clear()
print(x)
```

```python
len(person)
for value in person.values():
    print(value)
```

```python
len(person)
for keys in person.keys():
    print(keys)
```

```python
len(person)
for key,value in person.items():
    print(f"{key}:{value}")
```

```python
person["is_alive"]=True
len(person)
for key,value in person.items():
    print(f"{key}:{value}")
```

```python
import random
people = [
    {
        "name": "Ram",
        "address": "KTM",
        "age": random.randint(20, 50),
        "salary": round(random.uniform(30000, 60000), 2),
        "is_alive": random.choice([True, False])
    },
    {
        "name": "Sita",
        "address": "PKR",
        "age": random.randint(20, 50),
        "salary": round(random.uniform(30000, 60000), 2),
        "is_alive": random.choice([True, False])
    },
    {
        "name": "Hari",
        "address": "BKT",
        "age": random.randint(20, 50),
        "salary": round(random.uniform(30000, 60000), 2),
        "is_alive": random.choice([True, False])
    },
    {
        "name": "Gita",
        "address": "Dharan",
        "age": random.randint(20, 50),
        "salary": round(random.uniform(30000, 60000), 2),
        "is_alive": random.choice([True, False])
    },
    {
        "name": "Shyam",
        "address": "Bhaktapur",
        "age": random.randint(20, 50),
        "salary": round(random.uniform(30000, 60000), 2),
        "is_alive": random.choice([True, False])
    }
    
]

# Loop through the list of people and print key-value pairs
i=0
for person in people:
    print(f"Person Details:")
    for key, value in person.items():
        print(f"{key}: {value}")
    print()
```

```python
keys=data.keys()
print(keys)
```
